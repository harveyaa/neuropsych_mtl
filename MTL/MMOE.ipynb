```python
import pandas as pd
import numpy as np
import torch.nn as nn
import torch
import os

from torch.utils.data import Subset
from torch.utils.data import DataLoader
```

```python
from miniMTL.datasets import caseControlDataset
from miniMTL.models import *
from miniMTL.training import Trainer
from miniMTL.mmoe import MMOEModel
```

```python
bs = 16
fold = 0
head = 3

# Define paths to data
p_pheno = os.path.join('/Users/harveyaa/Documents/masters/data','pheno_26-01-22.csv')
p_ids = '/Users/harveyaa/Documents/masters/neuropsych_mtl/datasets/cv_folds/intrasite'
p_conn = os.path.join('/Users/harveyaa/Documents/masters/data','connectomes')

# Create datasets
#cases = ['SZ','ASD','ADHD','BIP']
cases = ['DUP22q11_2','DEL15q11_2','DUP1q21_1','DEL1q21_1','DUP16p11_2','DEL16p11_2','DEL22q11_2','ADHD','ASD','SZ','BIP']
data = []
for case in cases:
    print(case)
    data.append(caseControlDataset(case,p_pheno,id_path=p_ids,conn_path=p_conn,
                                    type='conn',strategy='balanced',format=0))
    
# Split data & create loaders & loss fns
loss_fns = {}
trainloaders = {}
testloaders = {}
decoders = {}
for d, case in zip(data,cases):
    train_idx, test_idx = d.split_data(random=False,fold=fold,splits=(0.75,0.25))
    train_d = Subset(d,train_idx)
    test_d = Subset(d,test_idx)
    trainloaders[case] = DataLoader(train_d, batch_size=bs, shuffle=True)
    testloaders[case] = DataLoader(test_d, batch_size=bs, shuffle=True)
    loss_fns[case] = nn.CrossEntropyLoss()
    decoders[case] = eval(f'head{head}().double()')
```

```python
lr = 1e-3
num_epochs = 10

# Create model
model = MMOEModel(3, 3, decoders,loss_fns)
    
# Create optimizer & trainer
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
trainer = Trainer(optimizer,log_dir='.')
```

```python
p0 = []
for g in model.parameters():
    p0.append(g.clone())
```

```python
# Train model
trainer.fit(model,trainloaders,testloaders,num_epochs=num_epochs)

# Evaluate at end
metrics = model.score(testloaders)
for key in metrics.keys():
    print()
    print(key)
    print('Accuracy: ', metrics[key]['accuracy'])
    print('Loss: ', metrics[key]['loss'])
print()
```
